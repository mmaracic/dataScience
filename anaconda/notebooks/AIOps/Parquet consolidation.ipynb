```python
from os import listdir
from os.path import isfile, join
mypath = '/opt/notebooks/AIOps/span-parquet'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
print(onlyfiles)
```

```python
#read all data from files
import pandas as pd
data = pd.DataFrame()
consolidatedIndex = 0
for i in range(0, len(onlyfiles)):
    f = onlyfiles[i]
    if (i > 0 and i % 100 == 0):
        fileName = '/opt/notebooks/AIOps/parquet/all_data_' + str(consolidatedIndex) + '.parquet'
        data.to_parquet(fileName)
        print ('Written consolidated file ' + fileName)
        consolidatedIndex+=1
        data = pd.DataFrame()
    file = join(mypath, f)
    day = pd.read_parquet(file)
    data = data.append(day, ignore_index=True)
if (data.shape[0]>0):
    #final file
    fileName = '/opt/notebooks/AIOps/parquet/all_data_' + str(consolidatedIndex) + '.parquet'
    data.to_parquet(fileName)
    print ('Written consolidated file ' + fileName)
print('Done')
```
