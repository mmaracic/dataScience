```python
def isNan(value):
    return value != value
```

```python
def transformCrisp(values):
    dict = {}
    nonNullValues = []
    for i in range(0, values.size):
        value = values[i]
        if not isNan(value):
            nonNullValues.append(value)
        else:
            dict[value] = value 
    for i in range(0, len(nonNullValues)):
        value = nonNullValues[i]
        dict[value] = (2*i/(len(nonNullValues)-1))-1
    return dict
```
